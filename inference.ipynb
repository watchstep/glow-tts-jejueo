## Define TTS

```python
from korean.ko_cleaner import tokenize
import IPython
import os

tts_path="/content/drive/MyDrive/glow-tts-dialect/model/glow-tts/run-December-21-2022_08+28AM-0000000"
vocoder_path="/content/drive/MyDrive/glow-tts-dialect/model/hifigan/run-December-20-2022_03+12PM-0000000"
audio_path = "/content/drive/MyDrive/glow-tts-jejueo-demo/audio"
```

# Run Inference

```python
text = "혼저옵세예"
tokens = tokenize(text)
tokens = "".join(tokens)

!tts --text "$tokens" \
    --model_path "$tts_path"/best_model.pth \
    --config_path "$tts_path"/config.json \
    --out_path "$audio_path"/example1.wav \
    --vocoder_path "$vocoder_path"/best_model.pth \
    --vocoder_config_path "$vocoder_path"/config.json
```

```python
 # display audio
IPython.display.display(IPython.display.Audio(os.path.join(audio_path, "example1.wav")))
```
